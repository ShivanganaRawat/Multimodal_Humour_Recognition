```python
from glob import glob
import opensmile
import pandas as pd
```

```python
filepaths = sorted(glob(f"/content/drive/MyDrive/IR Project/M2H2-dataset/Main-Dataset/Raw-Audio/**/**/*.mp3"))
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.ComParE_2016,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors,
)
df = pd.DataFrame()
for filepath in filepaths:
  tempdf = smile.process_file(filepath)
  df = pd.concat([df, tempdf], axis=0)
df
```

```python
df.to_csv("/content/drive/MyDrive/IR Project/M2H2-dataset/Main-Dataset/Raw-Audio/audio_features.csv")
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/IR Project/M2H2-dataset/Main-Dataset/Raw-Audio/audio_features.csv")
df
```

```python
files = df.groupby('file')
avgdf = pd.DataFrame()
for filepath, group in files:
    group_mean = group.mean(axis=0)
    tempdf = pd.DataFrame(group_mean).T
    tempdf['file'] = filepath
    avgdf = pd.concat([avgdf, tempdf], axis=0)
avgdf
```

```python
episodes = []
scenes = []
filenames = []
for filepath in avgdf['file']:
  episode = filepath.split('/')[-3]
  scene = filepath.split('/')[-2]
  filename = filepath.split('/')[-1].split('.')[0]
  episodes.append(episode)
  scenes.append(scene)
  filenames.append(filename)
avgdf['episode'] = episodes
avgdf['scene'] = scenes
avgdf['filename'] = filenames
avgdf['ep_scene_utter'] = avgdf['episode'] + avgdf['scene'] + avgdf['filename']
avgdf
```
